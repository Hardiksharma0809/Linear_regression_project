### Data Analysis

```python
df.describe()
```

```python
plt.figure(figsize=(4,4))
sns.pairplot(df)
```

##### Age Distribution 

```python
sns.set()
plt.figure(figsize=(4,4))
sns.histplot(df["age"],kde = True)
plt.title("Age Distribution")
```

#### Sex Distribution

```python
sns.set()
plt.figure(figsize=(4,4))
sns.countplot(x = "sex" , data = df )
```

- There is not much difference in the count of Male and Female2

```python
df["sex"].value_counts().sum
```

#### BMI

```python
sns.set()
plt.figure(figsize=(4,4))
sns.histplot(df["bmi"],kde = True)
plt.title("BMI Distribution")
```

- Here we can see that BMI is normally distributed 

- Normal BMI range is between 18 to 25 anything lower or upper than this is considered to be underweight or overweight 

#### Children

```python
sns.set()
plt.figure(figsize=(4,4))
sns.countplot(x = "children" , data = df )
```

```python
df["children"].value_counts().sum
```

#### Smoker

```python
sns.set()
plt.figure(figsize=(4,4))
sns.countplot(x = "smoker" , data = df )
```

```python
df["smoker"].value_counts().sum
```

#### Region

```python
sns.set()
plt.figure(figsize=(4,4))
sns.countplot(x = "region" , data = df )
```

```python
df["region"].value_counts().sum
```

#### Charges 

```python
sns.set()
plt.figure(figsize=(4,4))
sns.histplot(df["expenses"],kde = True)
plt.title("expense")
```

## Multivariate Analysis

```python
sns.set()
plt.figure(figsize=(4,4))
sns.histplot(x="expenses",kde = True,hue = "sex", data = df)
plt.title("expense")
```

```python
sns.set()
plt.figure(figsize=(4,4))
sns.histplot(x="expenses",kde = True,hue = "region", data = df)
plt.title("expense")
```

#### Age and Expenses

```python
sns.scatterplot(x = "expenses", y = "age",data = df )
```

```python
sns.scatterplot(x = "expenses", y = "bmi",data = df )
```

- There is a slight positive relationship between expenses and BMI
